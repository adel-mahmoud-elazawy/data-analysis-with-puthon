```python
def detect_outliers (df,column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound,upper_bound
```

```python
lower_income, upper_income = detect_outliers(df, "Annual_Income")
print(df["Annual_Income"].describe())
print((df["Annual_Income"] < 0).sum())
print(df["Annual_Income"].min(), df["Annual_Income"].max())
df["log_Annual_Income"] = np.log(df["Annual_Income"])
```

```python
lower_bound, upper_bound = detect_outliers(df, "Price_($)")
print(df["Price_($)"].describe())
print((df["Price_($)"] < 0).sum())
print(df["Price_($)"].min(), df["Price_($)"].max())
df["log_price"] = np.log(df["Price_($)"])
lower_log, upper_log = detect_outliers(df, "log_price")
print(f"lower bound is : {lower_log} and upper bound is {upper_log}")
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.boxplot(df["Annual_Income"], vert=False)
plt.title("Annual_Income before transformation")
plt.xlabel("Annual_Income")

plt.subplot(1,2,2)
plt.boxplot(df["log_Annual_Income"], vert=False)
plt.title("Annual_Income after transformation")
plt.xlabel("Annual_Income")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.boxplot(df["Price_($)"], vert=False)
plt.title("price before transformation")
plt.xlabel("price")

plt.subplot(1,2,2)
plt.boxplot(df["log_price"], vert=False)
plt.title("price after transformation")
plt.xlabel("price")

plt.tight_layout()
plt.show()
```

```python
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month

bins = [0,10000,25000,50000,100000]
labels = ["low","medium","high","premium"]
df["price_category"] = pd.cut(df["Price_($)"], bins=bins, labels=labels)

plt.figure(figsize=(8,3))
plt.boxplot(
    [df[df["price_category"] == category]["Price_($)"] for category in labels],
    tick_labels=labels,
    patch_artist=True,
    boxprops=dict(facecolor='skyblue')
)
plt.title("Price distribution across categories")
plt.xlabel("Price category")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for category in labels:
    plt.hist(
        df[df["price_category"] == category]["log_price"], 
        bins=20, 
        alpha=0.6, 
        label=category
    )

plt.title('Log Transformed Price Distribution by Price Category', fontsize=16)
plt.xlabel("Log Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(title='Price Category', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
category_count = df["price_category"].value_counts()

plt.bar(category_count.index, category_count, color='slateblue', edgecolor='black')
plt.title("Count of Cars By Price Category", fontsize=16, weight='bold')
plt.xlabel('Price Category')
plt.ylabel('Count')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
company_price = df.groupby("Company", observed=True)["Price_($)"].mean().sort_values()

plt.figure(figsize=(12,6))
company_price.plot(kind="bar", color="royalblue", edgecolor="black")
plt.title("Average Price by Company", fontsize=16, weight="bold")
plt.xlabel("Company")
plt.ylabel("Average Price ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
